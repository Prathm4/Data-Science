```python
ds1=ds.drop(['AT','AP','AH','AFDP','GTEP',
                      'TAT','CO','NOX'],axis=1)
ds1
```

```python
ds3=ds1.values
```

```python
ds3
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
scaler=preprocessing.StandardScaler()
ds4=scaler.fit_transform(ds3)
```

```python
ds4
```

```python
ds5=pd.DataFrame(ds4)
ds5
```

```python
x=ds4[:,[0,2]]
x
```

```python
x.shape
```

```python
y=ds4[:,1]
y
```

```python
y.shape
```

```python
model=Sequential()
model.add(Dense(10,input_dim=2,kernel_initializer='uniform',activation='linear'))
model.add(Dense(5,kernel_initializer='uniform',activation='linear'))
model.add(Dense(1,kernel_initializer='uniform',activation='relu'))
```

```python
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
history=model.fit(x,y,validation_split=0.33,epochs=150,batch_size=14)
```

```python
scores=model.evaluate(x,y)
print(scores)
```

```python
model.history.history.keys()
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
